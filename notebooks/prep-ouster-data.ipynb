```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
sensor_range_unit = 0.001
depth_image = cv2.imread("/tmp/latest_range.tif", cv2.IMREAD_ANYDEPTH).astype(np.double) * sensor_range_unit
ir_image = cv2.imread("/tmp/latest_ir.tif", cv2.IMREAD_ANYDEPTH)
```

```python
plt.figure(figsize=(60,20))
plt.imshow(depth_image)
```

```python
plt.figure(figsize=(60,20))
plt.imshow(ir_image, cmap="gray")
```
